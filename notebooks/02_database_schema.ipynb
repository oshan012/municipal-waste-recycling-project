```mysql
CREATE TABLE IF NOT EXISTS income_groups(
    income_id INT NOT NULL,
    income_level VARCHAR(20) NOT NULL,

    PRIMARY KEY(income_id)
    );
```

```mysql
CREATE TABLE IF NOT EXISTS countries(
    country_id INT NOT NULL,
    country_name VARCHAR(100) NOT NULL,
    region VARCHAR(50) NOT NULL,
    income_id INT NOT NULL,

    PRIMARY KEY(country_id)
    FOREIGN KEY(income_id) REFERENCES income_groups(income_id)
    );
```

```mysql
CREATE TABLE IF NOT EXISTS waste_production(
    waste_id INT NOT NULL,
    country_id INT NOT NULL,
    production_year YEAR,
    trash_per_capita FLOAT NOT NULL,

    PRIMARY KEY(waste_id)
    FOREIGN KEY(country_id) REFERENCES countries(country_id)
    );
```

```mysql
CREATE TABLE IF NOT EXISTS recycling_rates(
    recycling_id INT NOT NULL,
    country_id INT NOT NULL,
    recycling_year YEAR,
    recycling_rate FLOAT NOT NULL,

    PRIMARY KEY(recycling_id)
    FOREIGN KEY(country_id) REFERENCES countries(country_id)
    );
```
